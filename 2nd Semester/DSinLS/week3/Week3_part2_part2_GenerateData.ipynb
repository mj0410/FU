```
import pdfx

pdf_1 = pdfx.PDFx('part2_paper1.pdf')
metadata_1 = pdf_1.get_metadata()

pdf_2 = pdfx.PDFx('part2_paper2.pdf')
metadata_2 = pdf_2.get_metadata()
```

```
metadata_2
```

```
def get_all_values(d):
    if isinstance(d, dict):
        for v in d.values():
            yield from get_all_values(v)
    elif isinstance(d, list):
        for v in d:
            yield from get_all_values(v)
    else:
        yield d

values_list_1 = list(get_all_values(metadata_1))
values_list_2 = list(get_all_values(metadata_2))
```

```
id_1 = []
id_2 = []

for v in values_list_1 :
  if type(v) == str :
    if 'uuid' in v :
      id_1.append(v)

for v in values_list_2 :
  if type(v) == str :
    if 'uuid' in v :
      id_2.append(v)

docID_1 = id_1[0].lstrip('uuid:')
docID_2 = id_2[0].lstrip('uuid:')
```

### Convert metadata to dataframe

```
import pandas as pd

add_meta_df_1 = pd.DataFrame.from_dict(metadata_1, orient='index')
add_meta_df_2 = pd.DataFrame.from_dict(metadata_2, orient='index')
#add_meta_df_3 = pd.DataFrame.from_dict(metadata_3, orient='index')

add_meta_df_1 = add_meta_df_1.transpose()
add_meta_df_2 = add_meta_df_2.transpose()
#add_meta_df_3 = add_meta_df_3.transpose()
```

#### Organize new metadata


```
add_meta_df_1
```

```
add_meta_df_1[['journal','DOI']] = add_meta_df_1.Subject.str.split(",", expand=True)
sha = [docID_1]
add_meta_df_1['sha'] = sha
add_meta_df_1 = add_meta_df_1.rename(columns={'Title':'title', 'Author':'authors'})
add_meta_df_1 = add_meta_df_1[['sha', 'title', 'authors', 'doi', 'journal']]
add_meta_df_1
```

```
add_meta_df_2
```

```
doi = []
for v in values_list_2 :
  if type(v) == str :
    if 'doi' in v :
      doi = v
      quit

doi = doi.lstrip('https://doi.org/')
doi
```

```
add_meta_df_2[['journal','etc']] = add_meta_df_2.Subject.str.split(",", expand=True)
sha = [docID_2]

add_meta_df_2['sha'] = sha
add_meta_df_2['doi'] = doi

add_meta_df_2 = add_meta_df_2.rename(columns={'Title':'title', 'Author':'authors'})
add_meta_df_2 = add_meta_df_2[['sha', 'title', 'authors', 'doi', 'journal']]
add_meta_df_2
```

```
add_meta_df = add_meta_df_1.append(add_meta_df_2, ignore_index = True)
add_meta_df
```

```
text_1 = pdf_1.get_text()
text_2 = pdf_2.get_text()
```

```
file1 = open("paper1.txt","w")
  
# \n is placed to indicate EOL (End of Line) 
file1.write(text_1) 
file1.close()
```

```
file2 = open("paper2.txt","w")
  
# \n is placed to indicate EOL (End of Line) 
file2.write(text_2) 
file2.close()
```

```
with open('paper1.txt', 'r') as f1:
    t1 = f1.readlines()

with open('paper2.txt', 'r') as f2:
    t2 = f2.readlines()
```

```
print("paper1 text length : ", len(t1))
print("paper2 text length : ", len(t2))
```

```
abstract_1 = t1[30:36]
abstract_2 = t2[20:30]
```

```
body_text_1 = t1[38:829]
body_text_2 = t2[31:658]
```

```
def listToDict(lst):
  text_lst = []
  for i in lst :
    op = { 'text' : i }
    text_lst.append(op)
  return text_lst
```

```
dict_abs_1 = listToDict(abstract_1)
dict_body_1 = listToDict(body_text_1)

dict_abs_2 = listToDict(abstract_2)
dict_body_2 = listToDict(body_text_2)
```

```
import json

json1 = {'paper_id' : docID_1,
         'authors' : {"first":add_meta_df_1['author'][0]},
         'abstract' : dict_abs_1,
         'body_text' : dict_body_1}

json2 = {'paper_id' : docID_2,
         'authors' : {"first":add_meta_df_2['author'][0]},
         'abstract' : dict_abs_2,
         'body_text' : dict_body_2}

with open('paper1.json', 'w') as outfile:
  json.dump(json1, outfile)

with open('paper2.json', 'w') as outfile:
  json.dump(json2, outfile)
```
