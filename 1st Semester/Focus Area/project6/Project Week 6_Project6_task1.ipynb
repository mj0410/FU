###Import the dataset

####Load biom data

```
import biom
a = biom.load_table('67_otu_table.biom')
# convert the biom data to a data-frame and rename the index field
# this might takes some minutes for larger files
df= a.to_dataframe()
df= df.transpose()
df.index.names= ['#SampleID']
df.head()
```

####Load txt data

```
data = pd.read_csv('https://raw.githubusercontent.com/twbattaglia/MicrobeDS/master/data-raw/MovingPictures/550_prep_72_qiime_20161216-085930.txt', sep="\t", header=0) #,comment="#"

data.head()
```

####join the meta-data file(.txt) to the resulting data-frame

```
new_df = pd.merge(df, data, left_on='#SampleID', right_on='#SampleID', how='left')
new_df.head()
```

###Select Species

```
p = new_df.groupby('host_subject_id')
p.groups
```

```
p1 = new_df[new_df['host_subject_id'] == 'F4']
p2 = new_df[new_df['host_subject_id'] == 'M3']
```

```
p1_s = p1.groupby('common_sample_site')
p1_s.groups
```

```
F4_lp = p1[p1['common_sample_site'] == 'L_palm']
F4_rp = p1[p1['common_sample_site'] == 'R_palm']
F4_t = p1[p1['common_sample_site'] == 'Tongue']
F4_f = p1[p1['common_sample_site'] == 'feces']

M3_lp = p2[p2['common_sample_site'] == 'L_palm']
M3_rp = p2[p2['common_sample_site'] == 'R_palm']
M3_t = p2[p2['common_sample_site'] == 'Tongue']
M3_g = p2[p2['common_sample_site'] == 'feces']
```

Select microbiom 4396235, 4381553

```
#F4_LP_1 = F4_left_palm[['4479944', 'days_since_experiment_start']]

mrp_1 = M3_rp[['collection_timestamp', '4396235']]
mrp_2 = M3_rp[['collection_timestamp', '4381553']]
```

```
mrp_1 = mrp_1.rename(columns={'collection_timestamp': 'ds', '4396235': 'y'})

y_value=[]
for i in range(len(mrp_1)) :
  y_value.append(mrp_1.y[i])

ds = mrp_1['ds'].values.astype(object)

d = {'ds': ds, 'y': y_value}
new_mrp_1 = pd.DataFrame(data = d)

new_mrp_1.ds = pd.to_datetime(new_mrp_1.ds)
new_mrp_1 = new_mrp_1.sort_values(by='ds', ascending=True)
new_mrp_1.tail()
```

```
mrp_2 = mrp_2.rename(columns={'collection_timestamp': 'ds', '4381553': 'y'})

y_value_2=[]
for i in range(len(mrp_2)) :
  y_value_2.append(mrp_2.y[i])

ds = mrp_2['ds'].values.astype(object)

d = {'ds': ds, 'y': y_value_2}
new_mrp_2 = pd.DataFrame(data = d)

new_mrp_2.ds = pd.to_datetime(new_mrp_2.ds)
new_mrp_2 = new_mrp_2.sort_values(by='ds', ascending=True)
new_mrp_2.tail()
```

```
new_mrp_1_v = new_mrp_1.set_index('ds')
new_mrp_2_v = new_mrp_2.set_index('ds')
```

###Detect Outliers

```
fig, axes_out_delete = plt.subplots(1,2, figsize=(12,5))
sns.boxplot(y = "y", data=new_mrp_1, ax=axes_out_delete[0]).set(
    xlabel='4396235', 
    ylabel='Number of Samples',
    title = '(a)'
)
sns.boxplot(y = "y", data=new_mrp_2, ax=axes_out_delete[1]).set(
    xlabel='4381553', 
    ylabel='Number of Samples',
    title = '(b)'
)
```

```
Q1_1 = new_mrp_1.quantile(0.25)[0]
Q3_1 = new_mrp_1.quantile(0.75)[0]
IQR_1 = Q3_1 - Q1_1

new_mrp_1_ro = new_mrp_1[(new_mrp_1['y'] >= Q1_1-1.5*IQR_1) & (new_mrp_1['y'] <= Q3_1+1.5*IQR_1)]
new_mrp_1_ro
```

```
Q1_2 = new_mrp_2.quantile(0.25)[0]
Q3_2 = new_mrp_2.quantile(0.75)[0]
IQR_2 = Q3_2 - Q1_2

new_mrp_2_ro = new_mrp_2[(new_mrp_2['y'] >= Q1_2-1.5*IQR_2) & (new_mrp_2['y'] <= Q3_2+1.5*IQR_2)]
new_mrp_2_ro
```

###Compare additive seasonalities and multiplicative seasonalities

```
fig, axes_compare = plt.subplots(1,2, figsize=(20,6))

m = Prophet(seasonality_mode='multiplicative')
m.fit(new_mrp_1)
future = m.make_future_dataframe(periods=100)
forecast = m.predict(future)
fig1 = m.plot(forecast, ax=axes_compare[0])
axes_compare[0].set_title('(a)')

ms = Prophet()
ms.fit(new_mrp_1)
future_s = ms.make_future_dataframe(periods=100)
forecast_s = ms.predict(future_s)
fig2 = ms.plot(forecast_s, ax=axes_compare[1])
axes_compare[1].set_title('(b)')
```

```
unc_m = []
for i in range(len(forecast)) :
  unc_m.append(forecast['yhat_upper'][i]-forecast['yhat_lower'][i])
forecast['uncertainty'] = unc_m
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper', 'uncertainty']].tail(10)
```

```
unc = []
for i in range(len(forecast_s)) :
  unc.append(forecast_s['yhat_upper'][i]-forecast_s['yhat_lower'][i])
forecast_s['uncertainty'] = unc
forecast_s[['ds', 'yhat', 'yhat_lower', 'yhat_upper', 'uncertainty']].tail(10)
```

```
figmul1 = m.plot_components(forecast.tail(50))
```

```
figadd1 = ms.plot_components(forecast_s.tail(50))
```

```
fig, axes_compare2 = plt.subplots(1,2, figsize=(20,6))

m2 = Prophet(seasonality_mode='multiplicative')
m2.fit(new_mrp_2)
future_m2 = m2.make_future_dataframe(periods=60)
forecast_m2 = m2.predict(future_m2)
fig1 = m2.plot(forecast_m2, ax=axes_compare2[0])
axes_compare2[0].set_title('(a)')

ms2 = Prophet()
ms2.fit(new_mrp_2)
future_s2 = ms2.make_future_dataframe(periods=60)
forecast_s2 = ms2.predict(future_s2)
fig2 = ms2.plot(forecast_s2, ax=axes_compare2[1])
axes_compare2[1].set_title('(b)')
```

###change-point analysis

```
from fbprophet.plot import add_changepoints_to_plot

mc=Prophet()
mc.fit(new_mrp_1)
future_c = mc.make_future_dataframe(periods=60)
forecast_c = mc.predict(future_c)
fc1 = mc.plot(forecast_c)
ac1 = add_changepoints_to_plot(fc1.gca(), mc, forecast_c)
```
